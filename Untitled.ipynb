## 1.2. Краткий обзор обзор понятий теории вероятности
### Шаг 7. Дисперсия и моменты старших порядков
#### Задание

Случайная величина $ξ$ принимает значения $-1$, $0$, $1$ с вероятностями, соответственно равными $0.25$, $0.5$ и $0.25$.
Плотность распределения случайной величины $η$ принимает следующие значения:
* $f(x)=0$ при $x≤1$
* $f(x)=0.5$ при $1<x≤2$
* $f(x)=0$ при $2<x≤3$
* $f(x)=0.5$ при $3<x≤4$
* $f(x)=0$ при $x>4$

Найдите дисперсии $ξ$ и $η$.

**Решение**

Дисперсию дискретной случайной величины можно найти по формуле $D\xi = \sum_{i=1}^{\infty} (x_i-E\xi)^2 p_i$, однако, чаще используется формула $D\xi = E\xi^2 - (E\xi)^2$. Ранее мы выяснили, что мат. ожидание $E\xi=0$. Тогда, подставив значения в формулу получим $(-1-0)^2\times0.25+(1-0)^2\times0.25+(0-0)^2\times0.5=0.5$.

Дисперсия непрерывной случайной величины рассчитывается по формуле $D\xi = \int_{-\infty}^{\infty} (x-E\xi)^2 f(x) dx$. В нашём случае плотность не равна 0 в двух участках, поэтому вычислим дисперсию как сумму дисперсий по этим участкам.

Мат. ожидание уже подсчитано на предыдущем шаге $E\eta = 2.5$. Остаётся посчитать мат. ожидание квадрата случайной величины.  $D\eta=\int\limits_{-\infty}^\infty(x−E\eta)^2f(x)dx = \int\limits_{-\infty}^1(x−2.5)^2f(x)dx + \int\limits_{1}^2(x−2.5)^2f(x)dx + \int\limits_{2}^3(x−2.5)^2f(x)dx + \int\limits_{3}^4(x−2.5)^2f(x)dx + \int\limits_{4}^\infty(x−2.5)^2f(x)dx =\int\limits_{-\infty}^1 0dx + 0.5\int\limits_{1}^2(x−2.5)^2dx + \int\limits_{2}^3 0dx + 0.5\int\limits_{3}^4(x−2.5)^2dx + \int\limits_{4}^\infty 0dx = 0.5\int\limits_{1}^2(x−2.5)^2dx + 0.5\int\limits_{3}^4(x−2.5)^2dx = 0.5\int\limits_{1}^2(x^2 - 5x +\frac{25}{4})dx + 0.5\int\limits_{3}^4(x^2 - 5x +\frac{25}{4})dx = 0.5(\frac{x^3}{3} - \frac{5x^2}{2} + \frac{25x}{4})|_{1}^{2} + 0.5(\frac{x^3}{3} - \frac{5x^2}{2} + \frac{25x}{4})|_{3}^{4} = 0.5 * \frac{13}{12} + 0.5 * \frac{13}{12} = \frac{13}{12} = 1.08(3) \approx 1.1$
